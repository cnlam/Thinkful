```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Chi-square distribution

```python
# generate chi square distribution
dof = 8
chisquare = np.random.chisquare(dof, 100)

# plot histogram
plt.hist(chisquare, bins=20, color='b', alpha=0.35)

# add vertical line at the mean
plt.axvline(chisquare.mean(), color='b', linestyle='solid', linewidth=2)

# add vertical line lines at one standard deviation above/below the mean
plt.axvline(chisquare.mean() + chisquare.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chisquare.mean() - chisquare.std(), color='b', linestyle='dashed', linewidth=2)

# label axes
plt.xlabel('$\chi^{2}$', fontsize=15)
plt.ylabel('Frequency', fontsize=15)

plt.show()
```

## Geometric distribution

```python
# generate geometric distribution
p = 0.5
geometric = np.random.geometric(p, 100)

# plot histogram
plt.hist(geometric, bins=20, color='b', alpha=0.35)

# add vertical line at the mean
plt.axvline(geometric.mean(), color='b', linestyle='solid', linewidth=2)

# add vertical line lines at one standard deviation above/below the mean
plt.axvline(geometric.mean() + geometric.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(geometric.mean() - geometric.std(), color='b', linestyle='dashed', linewidth=2)

# label axes
plt.xlabel('random variable $x$', fontsize=15)
plt.ylabel('Frequency', fontsize=15)

plt.show()
```

## Laplace distribution

```python
# generate Laplace distribution
loc = 0
scale = 1
laplace = np.random.laplace(loc, scale, 100)

# plot histogram
plt.hist(laplace, bins=20, color='b', alpha=0.35)

# add vertical line at the mean
plt.axvline(laplace.mean(), color='b', linestyle='solid', linewidth=2)

# add vertical line lines at one standard deviation above/below the mean
plt.axvline(laplace.mean() + laplace.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(laplace.mean() - laplace.std(), color='b', linestyle='dashed', linewidth=2)

# label axes
plt.xlabel('$x$', fontsize=15)
plt.ylabel('Frequency', fontsize=15)

plt.show()
```
